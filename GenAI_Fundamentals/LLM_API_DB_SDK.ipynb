### Creating a Workspace Client

```python
from databricks.sdk import WorkspaceClient
from databricks.sdk.service.serving import ChatMessage, ChatMessageRole

w = WorkspaceClient()
```

### Sending a Chat Completion Request to the LLM

```python
response = w.serving_endpoints.query(
    name="databricks-claude-sonnet-4-5",
    messages = [
        ChatMessage(role=ChatMessageRole.SYSTEM, content="You are batman - the protector of Gotham City."),
        ChatMessage(role=ChatMessageRole.USER, content="How's Gotham City doing today?")
    ],
    max_tokens = 1024
)

print(f"RESPONSE:\n{response.choices[0].message.content}")
```
